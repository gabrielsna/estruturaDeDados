```python
# prompt:    -Acrescente os campos, comissão, vendedor e total de vendas;
#       -Exiba quantas linhas existem na planilha;
#       -Traga o valor da maior venda;
#       -Traga o valor da menor venda;
#       -Exiba somente as colunas vendedor, comissão e total de vendas;
#       -Dessas colunas, exiba somente as 4 ultimas linhas;
#       -Exiba os todos os nomes de vendedores sem repetição;
#       -Exibe todas as vendas que possuam valor que 500;
#       -Informe um nome de vendedor e exiba todas as vendas;
#       -Exiba todas as vendas do vendedor "nome" maior que 500
#       -Plote um gráfico de vendas por vendedor

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
# drive.mount('/content/drive') # Mount Google Drive

# Biblioteca pandas
path = '/content/drive/MyDrive/testeED/arqExcel.csv'
dataFrame = pd.read_csv(path, delimiter = ';')

#print(dataFrame.head(3))
#print(dataFrame.tail(2))
#print(dataFrame["produto"][1])
#print(dataFrame["produto"])
#print(dataFrame[["produto", "valor"]])

# Adicionando colunas
dataFrame['comissão'] = [0.05, 0.1, 0.02, 0.08, 0.03, 0.07] # Adicione os valores corretos para comissão
dataFrame['vendedor'] = ['Ana', 'João', 'Maria', 'Pedro', 'Lucas', 'Julia'] # Adicione os valores corretos para vendedores
dataFrame['total_vendas'] = dataFrame['valor'] * (1 + dataFrame['comissão']) # Exemplo de cálculo, ajuste conforme necessário

# Número de linhas
num_linhas = len(dataFrame)
print(f"Número de linhas na planilha: {num_linhas}")

# Maior venda
maior_venda = dataFrame['valor'].max()
print(f"Maior venda: {maior_venda}")

# Menor venda
menor_venda = dataFrame['valor'].min()
print(f"Menor venda: {menor_venda}")

# Exibindo colunas vendedor, comissão e total de vendas (últimas 4 linhas)
print(dataFrame[['vendedor', 'comissão', 'total_vendas']].tail(4))

# Nomes de vendedores sem repetição
vendedores_unicos = dataFrame['vendedor'].unique()
print(f"Nomes de vendedores únicos: {vendedores_unicos}")

# Vendas com valor maior que 500
vendas_acima_500 = dataFrame[dataFrame['valor'] > 500]
print("\nVendas acima de 500:")
print(vendas_acima_500)

# Vendas de um vendedor específico
nome_vendedor = 'João' # Insira o nome do vendedor desejado
vendas_vendedor = dataFrame[dataFrame['vendedor'] == nome_vendedor]
print(f"\nVendas do vendedor {nome_vendedor}:")
print(vendas_vendedor)


# Vendas de um vendedor específico acima de 500
nome_vendedor = 'João'  # Substitua pelo nome do vendedor desejado
vendas_vendedor_acima_500 = dataFrame[(dataFrame['vendedor'] == nome_vendedor) & (dataFrame['valor'] > 500)]
print(f"\nVendas do vendedor {nome_vendedor} acima de 500:")
print(vendas_vendedor_acima_500)


# Gráfico de vendas por vendedor
plt.figure(figsize=(10, 6)) #Ajustar tamanho do gráfico
plt.bar(dataFrame['vendedor'], dataFrame['valor'])
plt.xlabel("Vendedor")
plt.ylabel("Valor da Venda")
plt.title("Vendas por Vendedor")
plt.show()
```
